```python
import numpy as np
import seaborn as sns
# Assuming 'Total Doses' is the column with positive skewness in cov_vac DataFrame
total_doses = cov_vac['Total Doses']

# Logarithmic transformation
transformed_total_doses = np.log1p(total_doses)

# Check the skewness of the transformed data
transformed_skewness = skew(transformed_total_doses)

print(f"Skewness of Transformed 'Total Doses': {transformed_skewness}")
# KDE plot of the transformed data
sns.distplot(transformed_total_doses,kde=True)
plt.title('Kernel Density Estimate (KDE) Plot: Transformed Data')
plt.xlabel('Transformed Values')
plt.ylabel('Density')
plt.show()
```

```python
#now we can say that it is normally distributed

```

```python
kurtosis_transformed_total_doses = kurtosis(transformed_total_doses)

print(f"Kurtosis of Transformed 'Total Doses': {kurtosis_transformed_total_doses}")
```

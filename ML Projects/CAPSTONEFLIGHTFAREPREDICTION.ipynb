```python
#feature Engineering
data["Jouney_day"]=pd.to_datetime(data.Date_of_Journey,format="%d/%m/%Y").dt.day
data["Jouney_Month"]=pd.to_datetime(data.Date_of_Journey,format="%d/%m/%Y").dt.month
```

```python
data.head()
```

```python
data["Dep_hr"]=pd.to_datetime(data.Dep_Time).dt.hour
data["Dep_min"]=pd.to_datetime(data.Dep_Time).dt.minute
```

```python
data.head()
```

```python
data["Arr_hr"]=pd.to_datetime(data.Arrival_Time).dt.hour
data["Arr_min"]=pd.to_datetime(data.Arrival_Time).dt.minute
data.head()
```

```python
data=data.drop(columns=["Dep_Time","Date_of_Journey","Arrival_Time"],inplace=True)
```

```python
data.head()
```

```python
duration = list(data.Duration)
duration_hr = []
duration_min = []

for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if 'h' in duration[i]:
            duration[i] = duration[i] + '0m'
        else:
            duration[i] = '0h' + duration[i]
    hours_part, minutes_part = duration[i].split('h')

    try:
        hours = int(hours_part)
    except ValueError:
        hours = 0

    try:
        minutes = int(minutes_part.split('m')[0])
    except ValueError:
        minutes = 0
    duration_hr.append(hours)
    duration_min.append(minutes)
```

```python
data["Duration_hr"]=duration_hr
data["Duration_min"]=duration_min
```

```python
data=data.drop("Duration",axis=1)
```

```python
data.head()
```

```python
data=data.drop(columns=["Route","Additional_Info"],axis=1)
```

```python
data.info()
```

```python
#Handling categorical values
category = []
for column in data.columns:
  if data[column].dtype =="object":
    category.append(column)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder().fit(data[category])
data[category] = oe.transform(data[category])
```

```python
import pandas as pd
data=pd.read_csv("/content/train.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.isna().sum()
```

```python
data=data.dropna()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
category = []
for column in data.columns:
  if data[column].dtype =="object":
    category.append(column)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder().fit(data[category])
data[category] = oe.transform(data[category])
```

```python
data.info()
```

```python
data.describe()
```

```python
x,y=data.drop("loss",axis=1),data["loss"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
import warnings
warnings.simplefilter("ignore")
```

```python
lr=LinearRegression().fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
ypred=lr.predict(x_test)
```

```python
ypred
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,ypred)
```

```python
lr.score(x,y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(y_test,ypred))
```
